```python
import pandas as pd
import numpy as np
```

```python
EPL_Data=pd.read_excel("EPL Match Data 2019-2023.xlsx")
```

```python
def set_points(f):
    points=0
    if f=='W':
        points=3
    elif f=='D':
        points=1
    else:
        points=0
    return points
```

```python
EPL_Data['Points']=EPL_Data['Result'].apply(set_points)
```
